```python
stocks = []
dates = []
BasicEPS = []
DilutedEPS = []
CashEPS = []
DivPS = []
OpRev = []
NPS = []
ROCE = []
NPMargin = []
ROA = []
ROE2Networth = []
NetIntMargin = []
#Cost2Income = []
#IntIncome2TotAst = []
#NonIntIncome2TotAst = []
#OpProf2TotAst = []
#OpEx2TotAst = []
#IntEx2TotAst = []
EntValue = []
EntValuePerNetSales = []
Price2Book = []
Price2Sales = []
RetentionRatios = []
EarningYield = []
AssetTurnRatio = []
EVPerEBITDA = []
PBDITMargin = []

for stock,URL in stock_list.items():
    print(stock)
    for page in pageNumbers:
        
        URL_d = URL.format(page)
        print(URL_d)
        time.sleep(10)
        
        page = requests.get(URL_d)

        soup = BeautifulSoup(page.content, "html.parser")

        job_elements = soup.find_all("tr")
        
        for i,tr in enumerate(soup.findAll('tr')):
            value = tr.text.strip().replace(',','').strip()
            
            if len(value) > 0:
                section = value.splitlines()
                for j, line in enumerate(section):
                    
                    if 'Key Financial Ratios' in line:
                        dates += section[j+1:j+6]
                    elif 'Basic EPS' in line:
                        BasicEPS += section[j+1:j+6]
                    elif 'Diluted EPS' in line:
                        DilutedEPS += section[j+1:j+6]
                    elif 'Cash EPS' in line:
                        CashEPS += section[j+1:j+6]
                    elif 'Operating Revenue / Share' in line or 'Revenue from Operations/Share' in line:
                        OpRev += section[j+1:j+6]
                    elif 'Dividend/Share' in line or 'Dividend / Share' in line:
                        DivPS += section[j+1:j+6]
                    elif 'Net Profit/Share' in line:
                        NPS += section[j+1:j+6]
                    elif 'ROCE' in line or 'Return on Capital Employed' in line:
                        ROCE += section[j+1:j+6]
                    elif 'Net Profit Margin' in line:
                        NPMargin += section[j+1:j+6]
                    elif 'Return on Assets' in line:
                        ROA += section[j+1:j+6]
                    elif 'Return on Equity / Networth' in line or 'Return on Networth / Equity' in line:
                        ROE2Networth += section[j+1:j+6]
#                    elif 'Cost to Income' in line:
#                        Cost2Income += section[j+1:j+6]
#                    elif 'Interest Income/Total Assets' in line:
#                        IntIncome2TotAst += section[j+1:j+6]
#                    elif 'Non Interest Income/Total Assets' in line:
#                        NonIntIncome2TotAst += section[j+1:j+6]
#                    elif 'Operating Profit/Total Assets' in line:
#                        OpProf2TotAst += section[j+1:j+6]
#                    elif 'Operating Expenses/Total Assets' in line:
#                        OpEx2TotAst += section[j+1:j+6]
#                    elif 'Interest Expenses/Total Assets' in line:
#                        IntEx2TotAst += section[j+1:j+6]
                    elif 'Enterprise Value' in line:
                        EntValue += section[j+1:j+6]
                    elif 'EV Per Net Sales' in line or 'EV/Net Operating Revenue' in line:
                        EntValuePerNetSales += section[j+1:j+6]
                    elif 'Price To Book Value' in line or 'Price/BV' in line:
                        Price2Book += section[j+1:j+6]
                    elif 'Price To Sales' in line or 'Price/Net Operating Revenue' in line:
                        Price2Sales += section[j+1:j+6]
                    elif 'Retention Ratios' in line:
                        RetentionRatios += section[j+1:j+6]
                    elif 'Earnings Yield' in line:
                        EarningYield += section[j+1:j+6]
                    
                    if 'Asset Turnover Ratio' in line:
                        AssetTurnRatio += section[j+1:j+6]
                    elif len(stocks) > len(AssetTurnRatio):
                        AssetTurnRatio += 5 * ['---']

                    if 'EV/EBITDA' in line:
                        EVPerEBITDA += section[j+1:j+6]
                    elif len(stocks) > len(EVPerEBITDA):
                        EVPerEBITDA += 5 * ['---']

                    if 'PBDIT Margin' in line:
                        PBDITMargin += section[j+1:j+6]
                    elif len(stocks) > len(PBDITMargin):
                        PBDITMargin += 5 * ['---']
                    
        stocks += 5 * [stock]
```
